```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from plotly import tools
```

```python
df = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/full_data.csv")
```

```python
country = df.location.unique()
```

```python
def cap_name(name):
  # Devuelve una string con todas las palabras con la primer letra en mayuscula
  texto = ""
  for i in name.split():
    texto = texto + i.capitalize() + " "
  return texto.rstrip()
```

<h1>Punto numero 1</h1>

```python
def get_name_from_user():
  # Pide un pais. Termina cuando se ingresa un pais valido
  while True:
    name = input("Ingrese un pais ")
    if cap_name(name) in country:
      break
  return cap_name(name)
```

```python
def plot_one_country():  
  # Grafica los casos totales y muertes totales de un pais
  name = get_name_from_user()
  filtered_data = df[df["location"]==name]

  fig = go.Figure(data=[
      go.Scatter(
          x=filtered_data.date,
          y=filtered_data.total_cases,
          name="Total Cases",
      ),
      go.Scatter(
          x=filtered_data.date,
          y=filtered_data.total_deaths,
          name="Total Deaths",
      )
  ])
  fig.update_layout(
    title="Covid-19 "+name,
    xaxis_title="Date",
    yaxis_title="Cases and Deaths",
)
  fig.show()
```

```python
# Para correr el punto numero 1, ejecutar esta celda.

plot_one_country()
```

<h1>Punto numero 2</h1>




```python
def get_two_countrys():
  # Pide ingresar dos paises. Termina cuando se ingresan dos paises validos.
  count = 0
  lista = []
  while True:
    name = input("Ingrese un pais ")
    if cap_name(name) in country:

      count = count + 1
      lista.append(cap_name(name))

      if count == 2:
        break

  return lista
```

```python
def plot_two_countrys():  
  # Grafica dos muertes totales y casos totales de dos paises.
  lista = get_two_countrys()

  data_country_1 = df[df["location"]==lista[0]]
  data_country_2 = df[df["location"]==lista[1]]

  fig = go.Figure(data=[
      
      # First
      go.Scatter(
          x=data_country_1.date,
          y=data_country_1.total_cases,
          name="Total Cases: " + lista[0],
          line=dict(color="red")

      ),
      go.Scatter(
          x=data_country_1.date,
          y=data_country_1.total_deaths,
          name="Total Deaths: " + lista[0],
          line=dict(color="red", dash='dot')
      ),
      
      # Second

      go.Scatter(
          x=data_country_2.date,
          y=data_country_2.total_cases,
          name="Total Cases: " + lista[1],
          line=dict(color="blue")
      ),
      go.Scatter(
          x=data_country_2.date,
          y=data_country_2.total_deaths,
          name="Total Deaths: " + lista[1],
          line=dict(color="blue", dash='dot')
      )
  ])

  fig.show()
```

```python
# Para correr el punto numero 2, ejecutar esta celda.

plot_two_countrys()
```

<h1>Punto numero 3</h3>


```python
def get_list_of_countrys():
  # Pide n ingresar n cantidad de paises. Termina cuando se ingresa la palabra "no"
  lista = []
  while True:
    name = input('Ingrese un pais. Ingrese "no" para salir')
    if cap_name(name) in country:
      lista.append(cap_name(name))

    if name.capitalize() == "No":
        break

  return lista
```

```python
def plot_n_countrys():
  # Grafica todos los paises que se pidieron.
  lista = get_list_of_countrys()

  fig = go.Figure(data=[])

  fig = tools.make_subplots(rows = 1, cols = 1)

  for i in range(0, len(lista)):
    data = df[df["location"]==lista[i]]
    trace_0 = go.Scatter(x=data.date,y=data.total_cases,name=lista[i] + ": Cases")
    trace_1 = go.Scatter(x=data.date,y=data.total_deaths,name= lista[i] + ": Deaths",line=dict(dash='dot'))
  
    fig.append_trace(trace_0,1,1)
    fig.append_trace(trace_1,1,1)

  fig.update_layout(
      xaxis=dict(
          rangeselector=dict(
          ),
          rangeslider=dict(
              visible=True
          ),
          type="date"
      )
  )
  fig.update_yaxes(type="log")
  fig.show()
```

```python
# Para correr el punto numero 3, ejecutar esta celda.

plot_n_countrys()
```
